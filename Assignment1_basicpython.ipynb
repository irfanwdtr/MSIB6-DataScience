```python
# SOAL 2
def calculate_investment(capital, tenure):
    # Cek apakah modal investasi dan jangka waktu investasi sesuai ketentuan
    if capital <= 0 or capital > 1000000000:
        print("Error: Modal investasi harus dalam rentang 1 hingga 1.000.000.000.")
        return
    if tenure < 1 or tenure > 25:
        print("Error: Jangka waktu investasi harus dalam rentang 1 hingga 25 tahun.")
        return

    # Hitung keuntungan setiap tahun
    total_investment = capital
    for year in range(1, tenure + 1):
        profit = total_investment * 0.25
        total_investment += profit
        print(f"Tahun ke- {year}, total investasi Anda: {total_investment}")

# Menerima input dari pengguna
try:
    print("==========================================================")
    print("----------------Investasi Jamin-Sugih Co------------------")
    print("==========================================================")
    capital = int(input("Modal Investasi               : "))
    tenure = int(input("Jangka Waktu Investasi (tahun): "))
    print("==========================================================")
    print("----------------------Hasil Inventasi---------------------")
    print("==========================================================")
    calculate_investment(capital, tenure)
except ValueError:
    print("Error: Masukkan harus berupa bilangan bulat.")
```
